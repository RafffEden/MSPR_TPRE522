```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
 
```

```python
img = cv.imread("Data/apple/Image_1.jpg",)
assert img is not None, "file could not be read, check with os.path.exists()"
blur = cv.blur(img,(50,50))
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
blur_rgb = cv.cvtColor(blur, cv.COLOR_BGR2RGB)
plt.subplot(121),plt.imshow(img_rgb, cmap = plt.cm.Spectral),plt.title('Original')
plt.xticks([]), plt.yticks([])
plt.subplot(122),plt.imshow(blur_rgb, cmap = plt.cm.Spectral),plt.title('Blurred')
plt.xticks([]), plt.yticks([])
plt.show()
```
